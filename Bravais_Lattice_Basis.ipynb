```python
# Import the necessary library
import lattpy as lp

# Define the basis vectors for each lattice
# Square
square_basis = [[1, 0], [0, 1]]

# Rectangular
rect_basis = [[2, 0], [0, 1]]

# Hexagonal
hex_basis = [[1, 0], [0.5, 0.866]]

# Centered Rectangular
centered_rect_basis = [[1, 0], [0.5, 0.5]]

# Oblique (example with angle 60 degrees)
import numpy as np
angle_rad = np.deg2rad(60)
oblique_basis = [[1, 0], [np.cos(angle_rad), np.sin(angle_rad)]]


# Create the lattices
square_lattice = lp.Lattice(square_basis)
rect_lattice = lp.Lattice(rect_basis)
hex_lattice = lp.Lattice(hex_basis)
centered_rect_lattice = lp.Lattice(centered_rect_basis)
oblique_lattice = lp.Lattice(oblique_basis)

# Add a minimal basis to each lattice for plotting purposes if not already done
if square_lattice.num_base == 0:
    square_lattice.add_atom([0, 0], element="A")
if rect_lattice.num_base == 0:
    rect_lattice.add_atom([0, 0], element="A")
if hex_lattice.num_base == 0:
     # Add basis atoms to the hexagonal lattice (e.g., graphene-like structure)
    hex_lattice.add_atom([0, 0], element="C")
    hex_lattice.add_atom([1/3, 1/3], element="C")
if centered_rect_lattice.num_base == 0:
    centered_rect_lattice.add_atom([0, 0], element="A")
if oblique_lattice.num_base == 0:
    oblique_lattice.add_atom([0, 0], element="A")


# You can now work with these lattice objects, for example, print some information
print("Square Lattice:")
print(square_lattice)
print("\nRectangular Lattice:")
print(rect_lattice)
print("\nHexagonal Lattice:")
print(hex_lattice)
print("\nCentered Rectangular Lattice:")
print(centered_rect_lattice)
print("\nOblique Lattice:")
print(oblique_lattice)
```

```python
import matplotlib.pyplot as plt

# Function to visualize a lattice
def plot_lattice(lattice, title):
    # Add connections and analyze based on lattice type
    if lattice.num_base == 1:
        # Add a connection from the single atom to itself (nearest neighbor)
        lattice.add_connection(0, 0, 1)
        lattice.analyze() # Analyze the lattice after adding connections
    elif lattice.num_base == 2: # Assuming hexagonal lattice with 2 basis atoms
         # Add nearest neighbor connections for a 2-atom basis (e.g., hexagonal)
         lattice.add_connection(0, 1, 1) # Connect atom 0 to atom 1 (nearest neighbor)
         lattice.add_connection(1, 0, 1) # Connect atom 1 to atom 0 (nearest neighbor)
         lattice.analyze() # Analyze the lattice after adding connections


    # Build the lattice to generate sites
    lattice.build((5, 5)) # Build a 5x5 grid of lattice sites

    lattice.plot(show=False)
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis('equal')
    plt.grid(True)
    plt.show()

# Visualize each lattice
plot_lattice(square_lattice, "Square Lattice")
plot_lattice(rect_lattice, "Rectangular Lattice")
plot_lattice(hex_lattice, "Hexagonal Lattice")
plot_lattice(centered_rect_lattice, "Centered Rectangular Lattice")
plot_lattice(oblique_lattice, "Oblique Lattice")
```

```python
import numpy as np

# 1. Simple Cubic (SC)
sc_basis = np.array([[1, 0, 0],
                     [0, 1, 0],
                     [0, 0, 1]])

# 2. Body-Centered Cubic (BCC)
bcc_basis = np.array([[-0.5, 0.5, 0.5],
                      [0.5, -0.5, 0.5],
                      [0.5, 0.5, -0.5]])

# 3. Face-Centered Cubic (FCC)
fcc_basis = np.array([[0, 0.5, 0.5],
                      [0.5, 0, 0.5],
                      [0.5, 0.5, 0]])

# 4. Simple Tetragonal (ST)
a, c = 1, 2 # Example values
st_basis = np.array([[a, 0, 0],
                     [0, a, 0],
                     [0, 0, c]])

# 5. Body-Centered Tetragonal (BCT)
a, c = 1, 2 # Example values
bct_basis = np.array([[-a/2, a/2, c/2],
                      [a/2, -a/2, c/2],
                      [a/2, a/2, -c/2]])

# 6. Simple Orthorhombic (SO)
a, b, c = 1, 2, 3 # Example values
so_basis = np.array([[a, 0, 0],
                     [0, b, 0],
                     [0, 0, c]])

# 7. Base-Centered Orthorhombic (BCO)
a, b, c = 1, 2, 3 # Example values
bco_basis = np.array([[a/2, b/2, 0],
                      [-a/2, b/2, 0],
                      [0, 0, c]])

# 8. Body-Centered Orthorhombic (BCOr) - Note: Name can vary, using BCOr for clarity
a, b, c = 1, 2, 3 # Example values
bcor_basis = np.array([[-a/2, b/2, c/2],
                       [a/2, -b/2, c/2],
                       [a/2, b/2, -c/2]])

# 9. Face-Centered Orthorhombic (FCO)
a, b, c = 1, 2, 3 # Example values
fco_basis = np.array([[0, b/2, c/2],
                      [a/2, 0, c/2],
                      [a/2, b/2, 0]])

# 10. Simple Monoclinic (SM)
a, b, c, beta = 1, 2, 3, np.deg2rad(100) # Example values, beta is the angle between a and c
sm_basis = np.array([[a, 0, 0],
                     [0, b, 0],
                     [c * np.cos(beta), 0, c * np.sin(beta)]])

# 11. Base-Centered Monoclinic (BCM)
a, b, c, beta = 1, 2, 3, np.deg2rad(100) # Example values
bcm_basis = np.array([[a/2, b/2, 0],
                      [-a/2, b/2, 0],
                      [c * np.cos(beta), 0, c * np.sin(beta)]])

# 12. Simple Triclinic (STri)
a, b, c, alpha, beta, gamma = 1, 2, 3, np.deg2rad(80), np.deg2rad(90), np.deg2rad(70) # Example values
# Calculation based on https://en.wikipedia.org/wiki/Fractional_coordinates
v = np.sqrt(1 - np.cos(alpha)**2 - np.cos(beta)**2 - np.cos(gamma)**2 + 2 * np.cos(alpha) * np.cos(beta) * np.cos(gamma))
stri_basis = np.array([[a, 0, 0],
                      [b * np.cos(gamma), b * np.sin(gamma), 0],
                      [c * np.cos(beta), c * (np.cos(alpha) - np.cos(beta) * np.cos(gamma)) / np.sin(gamma), c * v / np.sin(gamma)]])

# 13. Simple Rhombohedral (SR) - Note: Can be represented by a simple hexagonal cell with specific angle
a, alpha = 1, np.deg2rad(80) # Example values
sr_basis = np.array([[a, 0, 0],
                     [a * np.cos(alpha), a * np.sin(alpha), 0],
                     [a * np.cos(alpha), a * (np.cos(alpha) - np.cos(alpha)**2) / np.sin(alpha), a * np.sqrt(1 - 3 * np.cos(alpha)**2 + 2 * np.cos(alpha)**3) / np.sin(alpha)]])

# 14. Simple Hexagonal (SH)
a, c = 1, 2 # Example values
sh_basis = np.array([[a, 0, 0],
                     [-a/2, a * np.sqrt(3)/2, 0],
                     [0, 0, c]])
```
