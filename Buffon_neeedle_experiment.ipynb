```python
import numpy as np #匯入numpy模組

d, l = 1, 1 #間距、針長=1
ntrials = 10000 #模擬10000次
x = d / 2 * np.random.random_sample(ntrials) #x=0到d距離除以2
theta = np.pi / 2 * np.random.random_sample(ntrials) #theta=numpy模組除以二

p = np.sum(x <= l / 2 * np.sin(theta)) / ntrials #p=np總和(x<=二分之l成np模組sintheta)
pexact = 2 * l / np.pi / d #pexact=針與現的相交機率
pi = 2 * l/ (p*d)
print(f"d = {d:g}, l = {l:g}")
print(f"Monte Carlo estimate: {p:.5f}, exact answer: {pexact:.5f}")
print(f"Estimated pi, {pi:.5f}")
```
