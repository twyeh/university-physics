## 計算電場

**Reasoning**:

I will define the appropriate Gaussian surface and use Gauss's Law to calculate the electric field, then store the results. I will calculate the electric field and store the resulting field strength and direction for points on the cylinder's surface.



```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Assuming the charged plane has a uniform surface charge density sigma
# Let Q be the total charge on the plane. The area of the plane is (2 * plane_half_side)**2.
# So, sigma = Q / (2 * plane_half_side)**2
# For simplicity, let's assume a surface charge density sigma, and the electric field E is given by E = sigma / (2 * epsilon_0)
# where epsilon_0 is the permittivity of free space.
# We don't have a value for Q or sigma, so we will express the electric field in terms of sigma and epsilon_0.
# The electric field due to an infinite charged plane is perpendicular to the plane and has a constant magnitude.
# Since the plane is in the z=0 plane, the electric field is in the z direction.

# For a finite plane, the electric field is not uniform and constant,
# especially near the edges. However, using Gauss's Law with a cylinder
# suggests we are likely meant to approximate the plane as infinite,
# or at least large enough compared to the cylinder's radius within the context of a simplified problem.
# For a finite plane, the calculation using Coulomb's Law would be a complex integral.
# Given the prompt mentioning "高斯定律或庫侖定律" and the simple geometry,
# it's highly probable that an approximation using Gauss's Law for an infinite plane is intended
# for demonstrating the principle, as calculating the field of a finite plane using Coulomb's Law
# is significantly more complex and would require numerical integration.

# Let's proceed with the approximation using Gauss's Law for an infinite plane for simplicity.
# The electric field E due to an infinite charged plane with surface charge density sigma is:
# E = sigma / (2 * epsilon_0) in the direction perpendicular to the plane.
# Assuming the charge density sigma is positive, the field points away from the plane.
# For z > 0, the field is in the +z direction. For z < 0, the field is in the -z direction.

# Let's define a symbolic constant for sigma / (2 * epsilon_0) for now.
# In a real scenario, epsilon_0 = 8.854e-12 F/m.
electric_field_magnitude_per_sigma = 1.0 # Representing 1 / (2 * epsilon_0) for simplicity in visualization

# Now, calculate the electric field on the surface of the cylinder.
# The cylinder's surface can be parameterized by (r*cos(theta), r*sin(theta), z) where r = cylinder_radius.
# We need to consider points on the curved surface and potentially the top and bottom caps if they are part of the Gaussian surface.
# However, the prompt asks for the field "on the cylinder surface". This implies the entire surface of the cylinder.

# Let's define points on the cylinder's surface.
# We can use a grid of theta and z values.
theta_values = np.linspace(0, 2 * np.pi, 50)
z_values = np.linspace(cylinder_center_z - cylinder_height / 2, cylinder_center_z + cylinder_height / 2, 50)

# Create a meshgrid for theta and z
theta_grid, z_grid = np.meshgrid(theta_values, z_values)

# Calculate the x, y, z coordinates of points on the cylinder's curved surface
x_surface = cylinder_radius * np.cos(theta_grid)
y_surface = cylinder_radius * np.sin(theta_grid)
z_surface = z_grid

# Now, calculate the electric field vector at each point (x_surface, y_surface, z_surface)
# Based on the infinite plane approximation, the electric field is purely in the z-direction.
# E_x = 0
# E_y = 0
# E_z = electric_field_magnitude_per_sigma if z > 0, and -electric_field_magnitude_per_sigma if z < 0.
# At z=0 (on the plane), the field is technically undefined in this simple model or often considered zero.
# For points on the cylinder surface at z=0, we need to handle this carefully. Since the cylinder
# is centered at z=0 and has height 8, it will have points at z=0 on its curved surface.
# Let's assume for points exactly on the plane (z=0), the electric field is symmetric and points equally in both +z and -z directions
# or we can exclude these points. Given the continuous nature of the cylinder surface,
# it's more practical to assign a value based on the limit as z approaches 0.
# However, the electric field of an infinite plane is discontinuous at z=0 in this simple model.

# A more physically accurate approach for points on the cylinder surface at z=0 would acknowledge that these points are on the plane.
# But the problem asks for the field on the cylinder surface *due to* the charged plane.
# Let's consider points slightly above and below the plane.
# For simplicity in visualization later, let's define the z-component of the electric field as:
E_z_surface = np.where(z_surface > 0, electric_field_magnitude_per_sigma, -electric_field_magnitude_per_sigma)
# Handle z=0 specifically if needed, e.g., E_z_surface[z_surface == 0] = 0, but for a continuous surface this might be visually awkward.
# Let's keep the step function for now, which is consistent with the infinite plane approximation.

# The electric field vector at each point on the curved surface is (E_x, E_y, E_z) = (0, 0, E_z_surface)
E_x_surface = np.zeros_like(x_surface)
E_y_surface = np.zeros_like(y_surface)

# Now, let's consider the top and bottom caps of the cylinder.
# Top cap: z = cylinder_center_z + cylinder_height / 2, with x^2 + y^2 <= cylinder_radius^2
z_top_cap = cylinder_center_z + cylinder_height / 2
# On the top cap, z > 0 (since cylinder_center_z=0 and cylinder_height=8, z_top_cap = 4).
# The electric field on the top cap is in the +z direction.
# E_field_top_cap = (0, 0, electric_field_magnitude_per_sigma)

# Bottom cap: z = cylinder_center_z - cylinder_height / 2, with x^2 + y^2 <= cylinder_radius^2
z_bottom_cap = cylinder_center_z - cylinder_height / 2
# On the bottom cap, z < 0 (z_bottom_cap = -4).
# The electric field on the bottom cap is in the -z direction.
# E_field_bottom_cap = (0, 0, -electric_field_magnitude_per_sigma)

# The problem asks for the electric field "on the cylinder surface", which includes the curved surface, top cap, and bottom cap.

# Store the results. We can store the coordinates and the electric field vector components.
# For the curved surface:
electric_field_data_curved = {
    'x': x_surface.flatten(),
    'y': y_surface.flatten(),
    'z': z_surface.flatten(),
    'Ex': E_x_surface.flatten(),
    'Ey': E_y_surface.flatten(),
    'Ez': E_z_surface.flatten()
}

# For the top cap:
# We need coordinates on the top cap. Use a grid in polar coordinates for the cap.
r_cap_values = np.linspace(0, cylinder_radius, 20)
theta_cap_values = np.linspace(0, 2 * np.pi, 40)
r_cap_grid, theta_cap_grid = np.meshgrid(r_cap_values, theta_cap_values)

x_top_cap = r_cap_grid * np.cos(theta_cap_grid)
y_top_cap = r_cap_grid * np.sin(theta_cap_grid)
z_top_cap_coords = np.full_like(x_top_cap, z_top_cap)

E_x_top_cap = np.zeros_like(x_top_cap)
E_y_top_cap = np.zeros_like(y_top_cap)
E_z_top_cap = np.full_like(x_top_cap, electric_field_magnitude_per_sigma) # Field is in +z direction

electric_field_data_top_cap = {
    'x': x_top_cap.flatten(),
    'y': y_top_cap.flatten(),
    'z': z_top_cap_coords.flatten(),
    'Ex': E_x_top_cap.flatten(),
    'Ey': E_y_top_cap.flatten(),
    'Ez': E_z_top_cap.flatten()
}

# For the bottom cap:
z_bottom_cap_coords = np.full_like(x_top_cap, z_bottom_cap) # Use same x, y grid as top cap

E_x_bottom_cap = np.zeros_like(x_top_cap)
E_y_bottom_cap = np.zeros_like(y_top_cap)
E_z_bottom_cap = np.full_like(x_top_cap, -electric_field_magnitude_per_sigma) # Field is in -z direction

electric_field_data_bottom_cap = {
    'x': x_top_cap.flatten(),
    'y': y_top_cap.flatten(),
    'z': z_bottom_cap_coords.flatten(),
    'Ex': E_x_bottom_cap.flatten(),
    'Ey': E_y_bottom_cap.flatten(),
    'Ez': E_z_bottom_cap.flatten()
}

# Combine data from all surfaces for storage
all_electric_field_data = {
    'x': np.concatenate((electric_field_data_curved['x'], electric_field_data_top_cap['x'], electric_field_data_bottom_cap['x'])),
    'y': np.concatenate((electric_field_data_curved['y'], electric_field_data_top_cap['y'], electric_field_data_bottom_cap['y'])),
    'z': np.concatenate((electric_field_data_curved['z'], electric_field_data_top_cap['z'], electric_field_data_bottom_cap['z'])),
    'Ex': np.concatenate((electric_field_data_curved['Ex'], electric_field_data_top_cap['Ex'], electric_field_data_bottom_cap['Ex'])),
    'Ey': np.concatenate((electric_field_data_curved['Ey'], electric_field_data_top_cap['Ey'], electric_field_data_bottom_cap['Ey'])),
    'Ez': np.concatenate((electric_field_data_curved['Ez'], electric_field_data_top_cap['Ez'], electric_field_data_bottom_cap['Ez']))
}

# Calculate the magnitude of the electric field at each point
all_electric_field_data['E_magnitude'] = np.sqrt(all_electric_field_data['Ex']**2 + all_electric_field_data['Ey']**2 + all_electric_field_data['Ez']**2)

# Display the first few points and their electric field
import pandas as pd
df_electric_field = pd.DataFrame(all_electric_field_data)
display(df_electric_field.head())
```

## 修改電場計算

### Subtask:
根據庫侖定律，定義計算有限大帶電平面電場的積分公式。


**Reasoning**:
The subtask requires defining the integral formula for the electric field of a finite charged plane using Coulomb's law. This involves setting up symbolic variables for the source and field points, the distance vector and its magnitude, the differential area element, and the integrand based on Coulomb's law. These steps can be combined into a single code block using SymPy.



```python
import sympy as smp

# 1. Define symbolic variables for the coordinates
# Source point (on the charged plane)
x_prime, y_prime, z_prime = smp.symbols('x_prime y_prime z_prime')
# Field point (on the cylinder surface) - These will be treated as constants during integration
x, y, z = smp.symbols('x y z')
# Surface charge density
sigma = smp.symbols('sigma')
# Permittivity of free space
epsilon_0 = smp.symbols('epsilon_0')

# For the charged plane in the z=0 plane, the source point has z_prime = 0.
z_prime = 0

# Source point vector
r_prime = smp.Matrix([x_prime, y_prime, z_prime])
# Field point vector (general point on cylinder surface)
r = smp.Matrix([x, y, z])

# 2. Define the vector from the source point to the field point
r_vec = r - r_prime

# 3. Define the magnitude of the distance between the source point and the field point
r_magnitude = smp.sqrt(r_vec.dot(r_vec))

# 4. Define the differential area element on the charged plane (in the x'-y' plane)
dA = smp.symbols('dA') # Symbolically represent the differential area element
# In Cartesian coordinates on the z'=0 plane, dA = dx' dy'
dx_prime, dy_prime = smp.symbols('dx_prime dy_prime')
dA = dx_prime * dy_prime

# 5. and 6. Define the integrand for the electric field contribution
# dE = (1 / (4 * pi * epsilon_0)) * (sigma * dA / r^2) * r_hat
# r_hat = r_vec / r_magnitude

# The integrand for the electric field vector components (dEx, dEy, dEz) is:
integrand_vec = (1 / (4 * smp.pi * epsilon_0)) * (sigma / r_magnitude**3) * r_vec

# The components of the integrand are:
integrand_Ex = integrand_vec[0]
integrand_Ey = integrand_vec[1]
integrand_Ez = integrand_vec[2]

# Display the components of the integrand
print("Integrand for Ex:")
display(integrand_Ex)
print("\nIntegrand for Ey:")
display(integrand_Ey)
print("\nIntegrand for Ez:")
display(integrand_Ez)

# 7. Define the limits of integration
# The charged plane is a square from x = -plane_half_side to +plane_half_side and y = -plane_half_side to +plane_half_side.
# The integration is over x_prime and y_prime.
# The limits for x_prime are from -plane_half_side to plane_half_side
# The limits for y_prime are from -plane_half_side to plane_half_side
# These limits will be used when performing the integration later.
x_prime_limits = (-plane_half_side, plane_half_side)
y_prime_limits = (-plane_half_side, plane_half_side)

print(f"\nIntegration limits for x_prime: {x_prime_limits}")
print(f"Integration limits for y_prime: {y_prime_limits}")
```

**Reasoning**:
The previous step defined the symbolic integrand for the electric field. The next logical step is to perform the integration of this integrand over the area of the charged plane to find the electric field at a given point (x, y, z). This requires numerical integration since the symbolic integration is likely complex. I will use `scipy.integrate.dblquad` to perform the double integration over x_prime and y_prime. I will define a function that takes the coordinates of the field point (x, y, z) and returns the components of the electric field vector (Ex, Ey, Ez).



```python
from scipy.integrate import dblquad
import numpy as np

# We need to perform the double integral of the integrand over the area of the plane.
# The integrand depends on the field point (x, y, z) and the source point (x_prime, y_prime).
# The integration is over x_prime from -plane_half_side to +plane_half_side and y_prime from -plane_half_side to +plane_half_side.

# Define the numerical value for epsilon_0 (permittivity of free space)
epsilon_0_val = 8.854e-12 # F/m

# We are given the total charge Q on the plane. The surface charge density sigma is Q / Area.
# Area = (2 * plane_half_side)**2
# Let's assume a value for Q for the calculation.
Q_val = 1.0 # Assuming a total charge of 1.0 unit (e.g., Coulomb)
plane_area = (2 * plane_half_side)**2
sigma_val = Q_val / plane_area

# Define the numerical integrands based on the symbolic expressions.
# These functions will take y_prime and x_prime as integration variables,
# and x, y, z, sigma_val, epsilon_0_val as fixed parameters.

def integrand_Ex_numerical(y_prime, x_prime, x, y, z, sigma_val, epsilon_0_val):
    # Based on the symbolic integrand: sigma*(x - x_prime)/(4*pi*epsilon_0*(z**2 + (x - x_prime)**2 + (y - y_prime)**2)**(3/2))
    r_sq = (x - x_prime)**2 + (y - y_prime)**2 + z**2
    # Avoid division by zero if the field point is on the plane (z=0) and directly above a source point (x=x_prime, y=y_prime).
    # For points on the cylinder surface, z is generally non-zero, or if z=0 on the curved surface, r_sq will be cylinder_radius^2 + (x-x_prime)^2 + (y-y_prime)^2 which is not zero.
    # However, numerically integrating near the plane requires care. Given the task is to plot on the cylinder *surface*,
    # we assume the field points are not exactly the same as the source points.
    if r_sq == 0:
        return 0 # Or handle as a special case depending on the physics context
    return (sigma_val * (x - x_prime)) / (4 * np.pi * epsilon_0_val * r_sq**(1.5))

def integrand_Ey_numerical(y_prime, x_prime, x, y, z, sigma_val, epsilon_0_val):
    # Based on the symbolic integrand: sigma*(y - y_prime)/(4*pi*epsilon_0*(z**2 + (x - x_prime)**2 + (y - y_prime)**2)**(3/2))
    r_sq = (x - x_prime)**2 + (y - y_prime)**2 + z**2
    if r_sq == 0:
        return 0
    return (sigma_val * (y - y_prime)) / (4 * np.pi * epsilon_0_val * r_sq**(1.5))

def integrand_Ez_numerical(y_prime, x_prime, x, y, z, sigma_val, epsilon_0_val):
    # Based on the symbolic integrand: sigma*z/(4*pi*epsilon_0*(z**2 + (x - x_prime)**2 + (y - y_prime)**2)**(3/2))
    r_sq = (x - x_prime)**2 + (y - y_prime)**2 + z**2
    if r_sq == 0:
        return 0
    return (sigma_val * z) / (4 * np.pi * epsilon_0_val * r_sq**(1.5))

# Define a function to calculate the electric field vector at a given field point (x, y, z)
def calculate_electric_field(x, y, z, plane_half_side, sigma_val, epsilon_0_val):
    # Integrate the numerical integrands over the plane's area (x_prime, y_prime)
    # dblquad(func, a, b, gfun, hfun, args=(), epsabs=1.49e-08, epsrel=1.49e-08)
    # func is the function to integrate, which takes y_prime then x_prime.
    # a, b are the limits for x_prime.
    # gfun, hfun are the limits for y_prime (can be functions of x_prime, but here they are constants).
    # args are the additional arguments to pass to func.

    # Integration limits for x_prime
    x_prime_lower = -plane_half_side
    x_prime_upper = plane_half_side

    # Integration limits for y_prime (constant limits)
    y_prime_lower = lambda x_prime: -plane_half_side
    y_prime_upper = lambda x_prime: plane_half_side

    # Perform the double integration for each component
    Ex, error_Ex = dblquad(integrand_Ex_numerical, x_prime_lower, x_prime_upper, y_prime_lower, y_prime_upper, args=(x, y, z, sigma_val, epsilon_0_val))
    Ey, error_Ey = dblquad(integrand_Ey_numerical, x_prime_lower, x_prime_upper, y_prime_lower, y_prime_upper, args=(x, y, z, sigma_val, epsilon_0_val))
    Ez, error_Ez = dblquad(integrand_Ez_numerical, x_prime_lower, x_prime_upper, y_prime_lower, y_prime_upper, args=(x, y, z, sigma_val, epsilon_0_val))

    return np.array([Ex, Ey, Ez])

# Example usage: Calculate the electric field at a point on the cylinder surface, e.g., top center of the top cap.
test_point_x = 0.0
test_point_y = 0.0
test_point_z = cylinder_center_z + cylinder_height / 2.0

E_at_test_point = calculate_electric_field(test_point_x, test_point_y, test_point_z, plane_half_side, sigma_val, epsilon_0_val)

print(f"Electric field at ({test_point_x}, {test_point_y}, {test_point_z}): {E_at_test_point}")
```
