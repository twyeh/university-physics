#題目

根據上述模型，分析圖中三個張力 $\vec{T}_1, \vec{T}_2, \vec{T}_3$ 的大小。

## 分析方法

利用 sympy 的線性方程式分析函數： linsolve ([等式集], (變數集))。

引進所有函數

```python
from sympy import *
```

宣告所有符號

```python
T1, T2, T3, M, g, alp, bet = symbols('T1 T2 T3 M g alp bet')
```

將三個等式寫成 linsolve() 函數可讀格式。

```python
sol=linsolve([Eq(-T1*cos(alp) + T2*cos(bet),0), Eq(T1*sin(alp)+ T2*sin(bet) - T3,0), Eq(T3-M*g,0)],(T1, T2, T3))
```

```python
sol
```

利用 trigsimp 簡化三角函數關係。

```python
trigsimp(sol)
```
