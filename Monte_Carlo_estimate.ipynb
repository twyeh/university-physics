```python
import numpy as np
d, l=1, 1
x=(d/2)*np.random.random_sample(1000)#random_sample生成0.0~1.0的隨機數字1000次{x|0<x<(d/2)}，其為針中心至線的距離。
theta=(np.pi/2)*np.random.random_sample(1000)#同理得{theta|0<theta<(π/2)}，其為針與線的夾角。
p=np.sum(x<=((l/2)*np.sin(theta)))/1000#當d=l時，sin(theta)超過針中心至線的距離時，其針與線會交叉，其為計算針與線交錯的針數。
print(f"d = {d}, l = {l}")
print(f"Monte Carlo estimate: {p:.5f}, exact answer: {((2*l)/(np.pi*d)):.5f}")#隨機抽樣估計的結果與理論值(p=2l/πd=2/π)比較。
#根據p=2/π的公式
pi=2/p
print(f"Monte Carlo estimate: {pi:.5f}, exact answer: {np.pi:.5f}")
```
