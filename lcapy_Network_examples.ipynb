```python
from lcapy import R
R1=R(1e2)
R2=R(2e2)
R3=R(3e2)
R4=R(4e2)
Rtot=(R1+R2)|(R3+R4)
Rtot.draw()
Rtot.R
```

```python
from lcapy import Circuit
Rtot=Circuit("""
P 0_1 0_2; right
W 0_1 0; right
W 0  1; up
W 1  4; down
R1 1 2 1e2; right
R2 2 3 2e2; right
R3 4 5 3e2; right
R4 5 6 4e2; right
R5 2 5 5e2; down
W 6 7; up
W 3 7; down
W 7 0_2; right
""")
Rtot.draw()
Rtot.simplify()
```

```python
from lcapy import Circuit
Rtot=Circuit("""
P 0_1 0_2;right
W 0_1 0; right
W 0  1; up
W 1  4; down
R1 1 2 1e2; right
R2 2 3 2e2; right
R3 4 5 3e2; right
R4 5 6 4e2; right
R6 2 5 6e2; down
W 6 7; up
W 3 7; down
W 7 0_2; right
W 0_1 8; down=2
V 8 9 9; right
W 9 0_2; up=2
W 0_1 10; up=2
R5 10 11 5e2; right
W 11 0_2; down=2
""")
Rtot.draw()
```
