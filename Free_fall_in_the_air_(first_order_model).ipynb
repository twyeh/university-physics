# Task
Design a program using `sympy` to calculate the velocity of a freely falling ball in air, assuming air resistance is proportional to velocity.

## Define symbols

### Subtask:
Define the necessary symbols using `sympy`.


**Reasoning**:
Import the `symbols` function from `sympy` and define the required symbols as real.



````python
from sympy import symbols

t, v, g, k = symbols('t v g k', real=True)
````

## Formulate the differential equation

### Subtask:
Write the differential equation for the falling ball with air resistance.


**Reasoning**:
Write the differential equation for the falling ball with air resistance, importing necessary functions and defining v as a function of t.



````python
from sympy import Function, Eq

v = Function('v')(t)
dv_dt = v.diff(t)
diff_eq = Eq(dv_dt, g - k*v)
````

## Solve the differential equation
```
## subtask
solve the differential equation using `sympy`'s differential equation solver.

### instructions
1. import the `dsolve` function from `sympy`.
2. use `dsolve` to solve the differential equation `diff eq` for the function `v(t)`.
3. store the solution in a variable.
```

### Subtask:
Solve the differential equation using `sympy`'s differential equation solver.


**Reasoning**:
Solve the differential equation using dsolve.



````python
from sympy import dsolve

solution = dsolve(diff_eq, v)
solution
````

## Apply initial conditions

### Subtask:
Apply the initial conditions to find the specific solution for the velocity.


**Reasoning**:
Apply the initial conditions to find the specific solution for the velocity by extracting the right-hand side of the general solution, defining the initial condition as an equation, solving for the integration constant, and substituting it back into the general solution.



````python
from sympy import solve

general_solution_rhs = solution.rhs
initial_condition = general_solution_rhs.subs(t, 0) - 0
C1 = solve(initial_condition, 'C1')[0]
specific_solution = general_solution_rhs.subs('C1', C1)
specific_solution
````

## Analyze the solution

### Subtask:
Display and analyze the velocity function obtained from the solution.


**Reasoning**:
Display the specific solution and describe its components.



````python
print("The specific solution for the velocity is:")
display(specific_solution)

print("\nAnalysis of the velocity function:")
print(f"The velocity function is given by v(t) = {specific_solution}")
print("The function has two terms: g/k and -g*exp(-k*t)/k.")
print("The term g/k is a constant term.")
print("The term -g*exp(-k*t)/k is an exponential term that decays to zero as time t increases.")
print("As time t approaches infinity, the exponential term exp(-k*t) approaches 0.")
print("Therefore, as t approaches infinity, the velocity v(t) approaches the constant term g/k.")
print("This constant term, g/k, represents the terminal velocity of the falling ball.")
````

## Summary:

### Data Analysis Key Findings

*   The differential equation for the velocity of a freely falling ball with air resistance proportional to velocity was successfully formulated as $\frac{dv}{dt} = g - kv$.
*   The general solution to the differential equation was found to be $v(t) = C\_1 e^{-kt} + \frac{g}{k}$.
*   Applying the initial condition $v(0) = 0$, the integration constant $C\_1$ was determined to be $-\frac{g}{k}$.
*   The specific solution for the velocity as a function of time was obtained as $v(t) = \frac{g}{k} - \frac{g e^{-kt}}{k}$.
*   The velocity function consists of a constant term, $\frac{g}{k}$, and a decaying exponential term, $-\frac{g e^{-kt}}{k}$.
*   As time approaches infinity, the exponential term approaches zero, and the velocity approaches the constant term $\frac{g}{k}$, which represents the terminal velocity.

### Insights or Next Steps

*   The derived velocity function can be used to calculate the velocity of the falling ball at any given time $t$, provided the values of $g$ and $k$ are known.
*   Further analysis could involve plotting the velocity function to visualize how the velocity changes over time and approaches terminal velocity.


# Task
Use sympy to derive and plot the velocity of a freely falling ball with air resistance proportional to velocity.

## Plot the velocity function

### Subtask:
Substitute numerical values for the symbols and plot the velocity function over time.


**Reasoning**:
Substitute numerical values into the specific solution and plot the resulting velocity function.



````python
from sympy import plot

g_val = 9.8
k_val = 0.1

velocity_numerical = specific_solution.subs({g: g_val, k: k_val})

p = plot(velocity_numerical, (t, 0, 50),
         title='Velocity of Falling Ball with Air Resistance',
         xlabel='Time (s)',
         ylabel='Velocity (m/s)',
         show=False)
p.show()
````

## Summary:

* The plot shows the velocity over time, ranging from 0 to 50 seconds.
* The plot is titled "Velocity of Falling Ball with Air Resistance", with the x-axis labeled "Time (s)" and the y-axis labeled "Velocity (m/s)".
