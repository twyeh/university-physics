```python
import matplotlib.pyplot as plt #繪製數據圖
import numpy as np #數據計算
from scipy.optimize import curve_fit #模型擬合分析


```

## 數據

距離數據 $xdata=[x1,x2,x3,\cdots,x10]$， 重量數據 $ytata=[y1,y2,y3,\cdots,y10]$。

```python
weightunit=9.8*0.0055*np.ones(10)
xdata=[0.0115,0.0120,0.0130,0.0140,0.0145,0.0150,0.0160,0.0180,0.0200,0.0215]
ydata=[11, 10, 9, 8, 7, 6, 5, 4, 3, 2]*weightunit
ydata
```

```python
xdata = np.asarray(xdata)
ydata = np.asarray(ydata)
#yerror = np.asarray(yerror)
#fig, ax = plt.subplots()
#plt.errorbar(xdata, ydata, yerr=yerror, xerr=0, fmt='none')
plt.plot(xdata, ydata, 'o')
plt.xlabel(r"$r$")
plt.ylabel(r"$F$")
plt.title("Magnetic Force between two Bucky Balls")
plt.show()
```

```python
# Define the Model function
def Model(x, A, B):
    y = A*x**B
    return y
```

```python
parameters, covariance = curve_fit(Model, xdata, ydata)
parameters[0],parameters[1]
```

```python
fit_A = parameters[0]
fit_B = parameters[1]
```

```python
fit_y = Model(xdata, fit_A, fit_B)
#plt.errorbar(xdata, ydata, yerr=yerror, xerr=0, fmt='none')
plt.plot(xdata, ydata, 'o', label='data')
plt.plot(xdata, fit_y, '-', label='fit')
plt.legend()
plt.xlabel(r"$r$")
plt.ylabel(r"$F$")
plt.title('')
plt.show()
```

```python
print("實驗模型為 A*r^B：")
print("A: {:.3g}".format(fit_A))
print( 'B: %5.2f' % fit_B)
```
