1. 假設有一個測試質點Q(質量為$m$,且位在座標 $Q=\vec{r}$),受到來自半圓環均質質量棒上的一個質點P(質量為$dM$,且位在座標 $P=\vec{r}^\prime$)的重力作用力
\begin{equation}
d\vec{F}_{PQ}=-G\frac{mdM}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)
\end{equation}

2. 由於質量棒為均質物體，其質量密度為定值，可以定義其質量密度為
$$\lambda=\frac{M}{l}=\frac{M}{\pi R}$$
其中 $M$ 為質量棒的總質量，$l=\pi R$ 為半圓形質量棒的總長度及$R$為半圓形棒的幾何半徑。

3. 根據上述定義，源質點$P$的質量$dM$可表示為 $$dM=d(\lambda l)=(d\lambda)l+\lambda dl=\lambda dl$$ 以及 $$dl=d(R\theta)=(dR)\theta + Rd\theta=Rd \theta $$ 其中利用弧長($l$)等於半徑($R$)乘以弧角($\theta$) $l=R\theta$ (半圓形時 $\theta=\pi$)。

4. P質點對Q質點所在位置的重力場等於
\begin{eqnarray}
d\vec{g}(\vec{r}) &=&\frac{d\vec{F}_{PQ}}{m}\\
&=&-G\frac{dM}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)\\
&=&-G\frac{\lambda dl}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)\\
&=&-G\frac{\lambda R d\theta}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)\\
\end{eqnarray}

5. 為簡化問題計算，平移坐標系，將原來座標原點平移至$Q$點的位置，此作用等於令$\vec{r}=0$:
\begin{eqnarray}
d\vec{g}(0)
&=&G\frac{\lambda R d\theta}{|\vec{r}^\prime|^3}(\vec{r}^\prime)\\
\end{eqnarray}
引進垂直坐標系,利用畢氏定理將向量$\vec{r}^\prime$表示成座標的分量 $\vec{r}^\prime=R(\cos\theta\hat{x}+\sin\theta \hat{y})$, $|\vec{r}^\prime|^3=R^3$
6. 現在將均勻棒上每一點對 $Q$ 的重力場的貢獻全部加起來，結果等於對 $d\vec{g}$ 進行積分:
\begin{eqnarray}
\vec{g}(0)&=&\int d\vec{g}(0) \\
&=&\int_0^{\pi} G\frac{\lambda R d\theta}{R^3}(R\cos\theta\hat{x}+R\sin\theta \hat{y})\\
&=&G\frac{\lambda}{R}\int_0^{\pi}d\theta(\cos\theta\hat{x}+\sin\theta \hat{y}) \\
&=&G\frac{\lambda}{R}(\int_0^{\pi}d\theta\cos\theta\hat{x}+\int_0^{\pi}d\theta\sin\theta \hat{y}) \\
&=&G\frac{\lambda}{R}(\sin\theta\hat{x}-\cos\theta \hat{y})_0^\pi \\
&=&G\frac{\lambda}{R}((\sin(\pi)-\sin(0))\hat{x}-(\cos\pi-\cos(0))\hat{y}) \\
&=&G\frac{\lambda}{R}((0-0)\hat{x}-((-1)-1))\hat{y}) \\
&=&G\frac{\lambda}{R}(0\hat{x}+2\hat{y}) \\
&=&2G\frac{\lambda}{R}\hat{y}
\end{eqnarray}

1. 假設有一個測試質點 $Q$ 位在座標原點，質量為 $m$。在球殼上任取一點$P$,質量為 $dM$, 則質點 $P$ 對質點 $Q$ 的重力為:
$$d\vec{F}_{PQ}(\vec{r}^\prime)=-G\frac{m dM}{|\vec{r}^\prime|^3}\vec{r}^\prime,$$
質點$P$對$Q$點的重力場為:
$$d\vec{g}(\vec{r}^\prime)=\frac{d\vec{F}_{PQ}(\vec{r}^\prime)}{m}=-G\frac{dM}{|\vec{r}^\prime|^3}\vec{r}^\prime,$$

2. $dM=\sigma da=\sigma (R\sin\theta d\theta)(R d\phi^\prime)$。其中小面積$da$可視為一個長為$R\sin\theta d\theta$及寬為$Rd\phi^\prime$的長方形的面積。在微小面積的微小極限近似下，面積可視為等同於一個點。這個經驗從高處向地面觀看在地面上平躺的一張紙時，該張紙幾乎變成一個白點。
$\sigma$為質量密度
$$\sigma=\frac{M}{4\pi R^2}$$ 其中 $R$ 為球殼的半徑，$4\pi$為球的立體角。 $4\pi R^2$ 球殼總面積。

3. 本題涉及複雜的座標標變換，為方便計算，可以選擇坐標系讓球殼的中心點$O$與$Q$點的連心線與座標系的$x$軸平行，並令$\overline{OQ}=r$,則$O$點的座標為 $O=-r\hat{x}$。

4. $P$點與$O$點間的向量差為
$$\vec{PO}=\vec{PQ}-\vec{OQ}=(\vec{r}^\prime-\vec{r})$$

5. 根據解答設定的圖示，若以$O$點為原點的坐標系，$P$點的座標可表示為
$$\vec{PO}=R(\cos\theta\hat{x}+\sin\theta\sin\phi^\prime\hat{y}+\sin\theta\cos\phi^\prime\hat{z})$$
其中 $\theta$ 為向量$\vec{PO}$與$x$軸間的夾角，$\phi^\prime$ 為$y-z$平面的水平角。

6. 將5中的以$O$為原點的座標系轉換成以$Q$為原點的座標系，$P$點的座標可表示成:
\begin{eqnarray}
\vec{PQ}&=&\vec{r}^\prime \\
    &=&\vec{r}+\vec{PO} \\
    &=&((R\cos\theta-r)\hat{x} \\
    & &+R\sin\theta\sin\phi^\prime\hat{y}\\
    & &+R\sin\theta\cos\phi^\prime\hat{z})
\end{eqnarray}

7. 利用重力場的疊加原理，球殼對$Q$點的重力場等於將$P$點對$Q$點的重力場循著球殼上每一點進行相加，此過程等同於對球殼面積進行積分:
\begin{eqnarray}
\vec{g}(\vec{r}^\prime)&=&\int d\vec{g}(\vec{r}^\prime)\\
&=&-\int G\frac{dM}{|\vec{r}^\prime|^3}\vec{r}^\prime \\
&=&- \int G\frac{\sigma da}{|\vec{r}^\prime|^3}\vec{r}^\prime \\
&=& -\int G\frac{\sigma R^2\sin\theta d\theta d\phi^\prime}{|\vec{r}^\prime|^3}\vec{r}^\prime \\
&=& -\int_0^{\pi}\sin\theta d\theta \int_0^{2\pi}d\phi^\prime G\frac{\sigma R^2}{|\vec{r}^\prime|^3}\vec{r}^\prime
\end{eqnarray}

8. 因為重力場$\vec{g}$是一個向量，需要針對每一個分量分別計算對應的積分。比較 $\vec{g}$與$\vec{r}^\prime$:
\begin{eqnarray}
\vec{g}&=&(g_x \hat{x}+ g_y \hat{y} + g_z \hat{z})\\
\vec{r}^\prime&=&((R\cos\theta-r)\hat{x}\\
&&+R\sin\theta\sin\phi^\prime\hat{y}\\
&&+R\sin\theta\cos\phi^\prime\hat{z})
\end{eqnarray}

9. 首先先計算 $g_x$:
\begin{eqnarray}
g_x(\vec{r}^\prime)=-\int_0^{\pi}\sin\theta d\theta \int_0^{2\pi}d\phi^\prime G\frac{\sigma R^2}{|\vec{r}^\prime|^3}(R\cos\theta-r)
\end{eqnarray}
其中 $$|\vec{r}^\prime|=\sqrt{R^2\sin^2\theta+(r-R\cos\theta)^2}。$$ 因為被積分函數中與$\phi^\prime$無關，可以先對其進行積分:
$$\int_0^{2\pi}d\phi^\prime=\phi^\prime|_0^{2\pi}=2\pi-0=2\pi$$ 前面的積分簡化成
\begin{eqnarray}
g_x(\vec{r}^\prime)=2\pi\sigma R^2\int_0^{\pi}\sin\theta d\theta \frac{r-R\cos\theta}{|\vec{r}^\prime|^3}
\end{eqnarray}
剩下的積分是一個複雜的積分， 一般都是查積分公式表。根據筆記給的積分公式
\begin{eqnarray}
\int_0^\pi \sin\theta d\theta \frac{(r-R\cos\theta)}{|\vec{r}^\prime |^3}
&=&-\int_1^{-1} d\cos\theta \frac{(r-R\cos\theta)}{|\vec{r}^\prime |^3}\\
&=&-\frac{2}{r^2}
\end{eqnarray}
其中利用變換變數法 $$\sin\theta d\theta = - d(\cos\theta)$$, 將對$\theta$的積分變換成對$\cos\theta$的積分，積分上下限則由 $\theta:\theta=0\to\theta=\pi$變成$\cos\theta:\cos(0)=1\to\cos(\pi=-1)$。多一個負號是因為 $d\cos\theta=-\sin\theta d\theta$。

10. 計算結果得到
$$g_x(\vec{r})=G(2\pi\sigma R^2)(-\frac{2}{r^2})=-G\frac{4\pi R^2 \sigma}{r^2}=-G\frac{M}{r^2}$$
