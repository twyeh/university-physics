1. Q($m$, $Q=\vec{r}$),P($dM$, $P=\vec{r}^\prime$)
\begin{equation}
d\vec{F}_{PQ}=-G\frac{mdM}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)
\end{equation}

2. 
$$\lambda=\frac{M}{l}=\frac{M}{\pi R}$$
 $M$ $l=\pi R$ $R$

3. $P$$dM$ $$dM=d(\lambda l)=(d\lambda)l+\lambda dl=\lambda dl$$  $$dl=d(R\theta)=(dR)\theta + Rd\theta=Rd \theta $$ ($l$)($R$)($\theta$) $l=R\theta$ ( $\theta=\pi$)

4. PQ
\begin{eqnarray}
d\vec{g}(\vec{r}) &=&\frac{d\vec{F}_{PQ}}{m}\\
&=&-G\frac{dM}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)\\
&=&-G\frac{\lambda dl}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)\\
&=&-G\frac{\lambda R d\theta}{|\vec{r}-\vec{r}^\prime|^3}(\vec{r}-\vec{r}^\prime)\\
\end{eqnarray}

5. $Q$$\vec{r}=0$:
\begin{eqnarray}
d\vec{g}(0)
&=&G\frac{\lambda R d\theta}{|\vec{r}^\prime|^3}(\vec{r}^\prime)\\
\end{eqnarray}
,$\vec{r}^\prime$ $\vec{r}^\prime=R(\cos\theta\hat{x}+\sin\theta \hat{y})$, $|\vec{r}^\prime|^3=R^3$
6.  $Q$  $d\vec{g}$ :
\begin{eqnarray}
\vec{g}(0)&=&\int d\vec{g}(0) \\
&=&\int_0^{\pi} G\frac{\lambda R d\theta}{R^3}(R\cos\theta\hat{x}+R\sin\theta \hat{y})\\
&=&G\frac{\lambda}{R}\int_0^{\pi}d\theta(\cos\theta\hat{x}+\sin\theta \hat{y}) \\
&=&G\frac{\lambda}{R}(\int_0^{\pi}d\theta\cos\theta\hat{x}+\int_0^{\pi}d\theta\sin\theta \hat{y}) \\
&=&G\frac{\lambda}{R}(\sin\theta\hat{x}-\cos\theta \hat{y})_0^\pi \\
&=&G\frac{\lambda}{R}((\sin(\pi)-\sin(0))\hat{x}-(\cos\pi-\cos(0))\hat{y}) \\
&=&G\frac{\lambda}{R}((0-0)\hat{x}-((-1)-1))\hat{y}) \\
&=&G\frac{\lambda}{R}(0\hat{x}+2\hat{y}) \\
&=&2G\frac{\lambda}{R}\hat{y}
\end{eqnarray}
