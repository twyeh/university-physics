## 定義參數和電荷位置

### Subtask:
設定電荷 Q 和距離參數 a，並定義兩個電荷在二維平面上的位置（例如：(-a, 0) 和 (a, 0)）。


**Reasoning**:
The subtask requires defining the charge Q, the distance parameter 'a', and the positions of two charges and a specific point in a 2D plane. I will use a code block to define these variables.



```python
import sympy

# Define the charge Q (Coulombs)
Q = 1e-9  # Example numerical value
# Q = sympy.Symbol('Q') # Uncomment to use a symbolic value

# Define the distance parameter 'a' (meters)
a = 1     # Example numerical value
# a = sympy.Symbol('a') # Uncomment to use a symbolic value

# Define the position of the first charge (-a, 0)
pos1 = (-a, 0)

# Define the position of the second charge (a, 0)
pos2 = (a, 0)

# Define the point to calculate the electric field (0, a)
point_to_calculate = (0, a)

print(f"Charge Q: {Q} C")
print(f"Distance parameter a: {a} m")
print(f"Position of charge 1: {pos1}")
print(f"Position of charge 2: {pos2}")
print(f"Point to calculate electric field: {point_to_calculate}")
```

## 計算特定點的電場

### Subtask:
計算距離兩個電荷皆為 \u221a2a 的點（例如：(0, a)）的電場。這將涉及使用庫侖定律計算每個電荷產生的電場，然後進行向量疊加。


**Reasoning**:
I need to calculate the electric field at a specific point (0, a) due to two charges. This involves importing the `math` module, defining Coulomb's constant, calculating position vectors, their magnitudes, and unit vectors, then using Coulomb's law to find the electric field contributed by each charge, and finally summing them vectorially. I will then print the components of the total electric field.



```python
import math

# Define Coulomb's constant (N m^2/C^2)
k = 8.9875e9

# Calculate the vector from pos1 to point_to_calculate
r1_vec = (point_to_calculate[0] - pos1[0], point_to_calculate[1] - pos1[1])

# Calculate the vector from pos2 to point_to_calculate
r2_vec = (point_to_calculate[0] - pos2[0], point_to_calculate[1] - pos2[1])

# Calculate the magnitude of these vectors
r1_mag = math.sqrt(r1_vec[0]**2 + r1_vec[1]**2)
r2_mag = math.sqrt(r2_vec[0]**2 + r2_vec[1]**2)

print(f"Vector from charge 1 to point: {r1_vec}")
print(f"Magnitude of vector r1: {r1_mag:.4f} m")
print(f"Verification: sqrt(2) * a = {math.sqrt(2) * a:.4f} m")
print(f"Vector from charge 2 to point: {r2_vec}")
print(f"Magnitude of vector r2: {r2_mag:.4f} m")
print(f"Verification: sqrt(2) * a = {math.sqrt(2) * a:.4f} m")

# Calculate the unit vector for each direction
r1_unit = (r1_vec[0] / r1_mag, r1_vec[1] / r1_mag)
r2_unit = (r2_vec[0] / r2_mag, r2_vec[1] / r2_mag)

print(f"Unit vector for r1: {r1_unit}")
print(f"Unit vector for r2: {r2_unit}")

# Calculate the electric field E1 (vector) produced by the charge at pos1
E1_magnitude = k * Q / (r1_mag**2)
E1 = (E1_magnitude * r1_unit[0], E1_magnitude * r1_unit[1])

# Calculate the electric field E2 (vector) produced by the charge at pos2
E2_magnitude = k * Q / (r2_mag**2)
E2 = (E2_magnitude * r2_unit[0], E2_magnitude * r2_unit[1])

print(f"Electric field E1: {E1} N/C")
print(f"Electric field E2: {E2} N/C")

# Sum the two electric field vectors to find the total electric field E_total
E_total = (E1[0] + E2[0], E1[1] + E2[1])

print(f"Total electric field E_total at {point_to_calculate}: {E_total} N/C")
```

## 生成二維電場分佈圖

### Subtask:
建立一個二維圖，以箭頭或流線圖表示電場向量，顯示電場強度分佈。圖中應標示出兩個電荷的位置、電場線方向以及圖例，以清楚展示電場的模式。


**Reasoning**:
I need to import `matplotlib.pyplot` and `numpy` for plotting and numerical operations. Then, I will define a function to calculate the electric field at any given point, encapsulating the physics logic from the previous step. This function will be crucial for calculating the field across a grid.



**Reasoning**:
I have already defined the `calculate_electric_field` function. Now I will generate a grid of points, calculate the electric field at each point using the defined function, and then visualize the electric field distribution using `streamplot` with charge locations, labels, title, and legend, as per the instructions.



**Reasoning**:
The previous code failed because `plt.streamplot` does not accept a `label` argument. I will remove the `label` argument from the `streamplot` function call and clarify the streamplot in the title or descriptive text instead of trying to add it to the legend directly.



```python
import matplotlib.pyplot as plt
import numpy as np

# Define Coulomb's constant (N m^2/C^2)
k = 8.9875e9

def calculate_electric_field(x, y, Q, pos1, pos2, k):
    """
    Calculates the total electric field (Ex, Ey) at a point (x, y)
    due to two charges Q at pos1 and pos2.
    """
    point = (x, y)

    # Electric field from charge 1
    r1_vec = (point[0] - pos1[0], point[1] - pos1[1])
    r1_mag_sq = r1_vec[0]**2 + r1_vec[1]**2
    # Avoid division by zero if point is exactly at the charge location
    if r1_mag_sq == 0:
        E1_x, E1_y = 0.0, 0.0  # Or handle as a singularity
    else:
        E1_magnitude = k * Q / r1_mag_sq
        r1_mag = np.sqrt(r1_mag_sq)
        r1_unit = (r1_vec[0] / r1_mag, r1_vec[1] / r1_mag)
        E1_x, E1_y = E1_magnitude * r1_unit[0], E1_magnitude * r1_unit[1]

    # Electric field from charge 2
    r2_vec = (point[0] - pos2[0], point[1] - pos2[1])
    r2_mag_sq = r2_vec[0]**2 + r2_vec[1]**2
    # Avoid division by zero if point is exactly at the charge location
    if r2_mag_sq == 0:
        E2_x, E2_y = 0.0, 0.0  # Or handle as a singularity
    else:
        E2_magnitude = k * Q / r2_mag_sq
        r2_mag = np.sqrt(r2_mag_sq)
        r2_unit = (r2_vec[0] / r2_mag, r2_vec[1] / r2_mag)
        E2_x, E2_y = E2_magnitude * r2_unit[0], E2_magnitude * r2_unit[1]

    # Sum the two electric field vectors
    E_total_x = E1_x + E2_x
    E_total_y = E1_y + E2_y

    return E_total_x, E_total_y


# Create a grid of points
# Use a range around the charges, for example, from -5*a to 5*a
x_range = np.linspace(-5 * a, 5 * a, 50)
y_range = np.linspace(-5 * a, 5 * a, 50)
X, Y = np.meshgrid(x_range, y_range)

# Initialize arrays to store electric field components
Ex = np.zeros_like(X)
Ey = np.zeros_like(Y)

# Calculate electric field at each point in the grid
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Ex[i, j], Ey[i, j] = calculate_electric_field(X[i, j], Y[i, j], Q, pos1, pos2, k)

# Create the plot
plt.figure(figsize=(10, 8))

# Plot electric field lines using streamplot (removed 'label' argument)
plt.streamplot(X, Y, Ex, Ey, color='blue', density=2, linewidth=1, arrowsize=1.5)

# Plot charge positions
plt.plot(pos1[0], pos1[1], 'ro', markersize=8, label=f'Charge 1 at ({pos1[0]}, {pos1[1]})') # Red circle for positive charge
plt.plot(pos2[0], pos2[1], 'ro', markersize=8, label=f'Charge 2 at ({pos2[0]}, {pos2[1]})') # Red circle for positive charge

# Add a dummy plot for the stream lines to appear in the legend
plt.plot([], [], color='blue', label='Electric Field Lines')

# Add labels, title, and legend
plt.xlabel('X-coordinate (m)')
plt.ylabel('Y-coordinate (m)')
plt.title('2D Electric Field Distribution for Two Positive Charges')
plt.legend()
plt.grid(True)

# Ensure aspect ratio is equal
plt.gca().set_aspect('equal', adjustable='box')

# Display the plot
plt.show()

print("2D electric field distribution plot generated.")
```
